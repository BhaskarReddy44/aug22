```python
def fibo():
    first_num=0
    yield first_num
    sec_num=1
    yield sec_num
    while(True):
        next_num=first_num+sec_num
        yield next_num
        first_num,sec_num=sec_num,next_num
g=fibo()
print(g)
#print(next(g))
#print(next(g))
#print(next(g))

for x in range(7):
    print(next(g))
```
